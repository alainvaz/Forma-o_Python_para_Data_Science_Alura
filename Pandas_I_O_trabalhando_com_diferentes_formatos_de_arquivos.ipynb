```python
from sqlalchemy import create_engine, MetaData, Table, inspect

# create_engine > conexao com banco de dados
# MetaData      > representa os metadados do banco de dados
# table         > representa a tabela do banco de dados
# inspect       > permite inspecionar o banco de dados
```

```python
sqlalchemy.__version__
```

```python
url_db = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url_db)
```

```python
dados.head()
```

```python
#criando engine > motor do banco

engine = create_engine('sqlite:///:memory')

```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
empregados = pd.read_sql(query, engine)
empregados
```

```python
query = 'SELECT * FROM clientes'
tudo = pd.read_sql(query, engine)
tudo
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
#ler tabela completa

pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```
